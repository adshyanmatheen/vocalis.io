```python
# Importing The Necessary Packages And Dependencies Required And Loading The Svarah Dataset For Data Exploration

import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from datasets import load_dataset
from huggingface_hub import login

warnings.filterwarnings('ignore')

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

plt.style.use('seaborn-v0_8-darkgrid')

login()

dataset = load_dataset('ai4bharat/Svarah')

print("The Available Splits In The Svarah Dataset:", dataset.keys())
print("")
print(dataset)
```

```python
# Creating A DataFrame For The Svarah Dataset, Generating An Overview Of The Svarah Dataset, And Performing Data Quality Validation

df = pd.DataFrame(dataset['test'])

print("")
print("Overview Of The Svarah Dataset")
print("")

print(f"Total Number Of Records : {len(df):,}")
print(f"Total Number Of Features: {len(df.columns)}")
print("")

print("The Names Of The Columns And The Types Of Data:")
print("")
display(df.dtypes)
print("")

print("The First Three Rows Of The Svarah Dataset")
print("")
display(df.head(3))
print("")

print("The Quality Validation Of The Svarah Dataset")
print("")

missing_counts = df.isnull().sum()

print(f"Missing Values: {missing_counts.sum()}")
print("")
print(f"Duplicate Records: {df.duplicated().sum()}")
print("")

if missing_counts.sum() > 0:
    print("The Missing Values Sorted By Column:")
    print("")
    print(missing_counts[missing_counts > 0])
    print("")
else:
    print("No Missing Values Was Found")
    print("")

print("The Shape Of The Svarah Dataset")
print("")
print(f"The Rows: {df.shape[0]:,}")
print("")
print(f"The Columns: {df.shape[1]}")
```

```python
# Performing Statistical And Visual Analysis Of The Audio Duration Distribution Amongst The Svarah Dataset

print("The Analysis Of The Audio Duration Within The Svarah Dataset")
print("")

print("The Statistics Of The Audio Duration: ")
print("")

print("")
print(f"The Total Number Of Hours : {df['duration'].sum() / 3600:.2f} Hours")
print(f"The Average Number Of Seconds: {df['duration'].mean():.2f} Seconds")
print(f"The Median Number Of Seconds: {df['duration'].median():.2f} Seconds")
print("")
print(f"The Minimum Number Of Seconds: {df['duration'].min():.2f} Seconds")
print(f"The Maximum Number Of Seconds: {df['duration'].max():.2f} Seconds")
print("")
print(f"The Standard Deviation Between Audio Recordings: {df['duration'].std():.2f} Seconds")
print(f"The First Quartile (25%) Of The Audio Recordings: {df['duration'].quantile(0.25):.2f} Seconds")
print(f"The Third Quartile (75%) Of The Audio Recordings: {df['duration'].quantile(0.75):.2f} seconds")
print("")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].set_title('The Distribution Of The Audio Duration Within The Svarah Dataset', fontweight='bold')

axes[0].set_xlabel('Duration In Seconds')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.3)

sns.boxplot( x=df['duration'], ax=axes[1], color='#2ecc71')

axes[1].set_title('The Audio Duration In Boxplot', fontweight='bold')

axes[1].set_xlabel('Duration In Seconds')
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Performing Statistical And Linguistic Analysis Of Text Transcriptions Within The Svarah Dataset

print("The Analysis Of Text Transcripts In The Svarah Dataset")
print("")

df['text_length'] = df['text'].astype(str).str.len()
df['word_count'] = df['text'].astype(str).str.split().str.len()
df['words_per_second'] = df['word_count'] / df['duration']

print("The Statistics Of The Text Transcripts: ")
print("")

print("")
print(f"The Average Length Of Text Transcripts: {df['text_length'].mean():.0f} Characters")
print(f"The Median Length Of Text Transcripts: {df['text_length'].median():.0f} Characters")
print(f"The Maximum Length Of Text Transcripts: {df['text_length'].max()} Characters")
print("")
print(f"The Minimum Length Of Text Transcripts: {df['text_length'].min()} Characters")
print(f"The Average Word Count Of Text Transcripts: {df['word_count'].mean():.0f} Words")
print(f"The Average Speech Rate Of Text Transcripts: {df['words_per_second'].mean():.2f} Words Per Second")
print("")

print("The Samples Of Text Transcripts: ")
print("")

for i in range(3):
    transcription = df['text'].iloc[i]

    if len(transcription) > 80:
        transcription = transcription[:80] + "..."

    print("")
    print(f"Sample Transcription {i + 1}")

    print("")
    print(f"Text Transcription : {transcription}")

    print("")
    print(
        f"The Duration Of The Audio Recordings: {df['duration'].iloc[i]:.2f} Seconds | "
        f"The Word Count Of The Text Transcripts: {df['word_count'].iloc[i]} | "
        f"The Rate Of Speech Of The Audio Recordings: {df['words_per_second'].iloc[i]:.2f} Word Per Second"
    )

print("")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(
    df['text_length'],
    bins=50,
    color='#9b59b6',
    edgecolor='black',
    alpha=0.75
)

axes[0].set_title('The Distribution Of Length In The Text Transcripts', fontweight='bold')

axes[0].set_xlabel('Number Of Characters')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.3)
axes[0].grid(axis='y', alpha=0.3)

axes[1].hist(
    df['word_count'],
    bins=50,
    color='#f39c12',
    edgecolor='black',
    alpha=0.75
)

axes[1].set_title('The Distribution Of Word Counts In Text Transcripts', fontweight='bold')

axes[1].set_xlabel('Number Of Words')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Performing Statistical Analysis And Visualization Of Gender Distribution Within The Svarah Dataset

print("The Distribution Of Gender In Svarah Dataset")
print("")

gender_distribution = df['gender'].value_counts()

print("")
print("The Counts Of Gender: ")
print("")
display(gender_distribution)
print("")

print("The Distribution Of Percentages: ")
print("")

gender_percentage_distribution = (gender_distribution / len(df) * 100).round(2)

for gender, percentage in gender_percentage_distribution.items():
    print(f"{gender:<10}: {percentage}%")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

gender_distribution.plot(
    kind='bar',
    ax=axes[0],
    color=['#e74c3c', '#3498db', '#2ecc71'],
    edgecolor='black',
    alpha=0.75
)

axes[0].set_title(
    'The Distributions Of Genders',
    fontweight='bold',
    fontsize=12
)

axes[0].set_xlabel('Gender Category')
axes[0].set_ylabel('Frequency Count')
axes[0].tick_params(axis='x', rotation=0)
axes[0].grid(axis='y', alpha=0.3)

axes[1].pie(
    gender_distribution,
    labels=gender_distribution.index,
    autopct='%1.1f%%',
    colors=['#e74c3c', '#3498db', '#2ecc71'],
    startangle=90
)

axes[1].set_title(
    'Gender Distribution Percentage',
    fontweight='bold',
    fontsize=12
)

plt.tight_layout()
plt.show()
```

```python
# Performing Statistical Analysis And Visualization Of Age Group Distribution Within The Svarah Dataset

print("The Distribution Of Age Groups In The Svarah Dataset")
print("")

age_group_distribution = (
    df['age-group']
    .value_counts()
    .sort_index()
)

print("")
print("The Counts Of Age Groups: ")
display(age_group_distribution)
print("")

print("")
print("The Distribution Of Percentages: ")

age_group_percentage_distribution = (age_group_distribution / len(df) * 100).round(2)

for age_group, percentage in age_group_percentage_distribution.items():
    print(f"{age_group:<15}: {percentage}%")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

age_group_distribution.plot(
    kind='bar',
    ax=axes[0],
    color='#3498db',
    edgecolor='black',
    alpha=0.75
)

axes[0].set_title(
    'The Distribution Of Age Groups',
    fontweight='bold',
    fontsize=12
)

axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Frequency Count')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

axes[1].pie(
    age_group_distribution,
    labels=age_group_distribution.index,
    autopct='%1.1f%%',
    startangle=90
)

axes[1].set_title(
    'The Distribution Percentage Of Age Groups',
    fontweight='bold',
    fontsize=12
)
```

```python
# Performing Statistical Analysis And Visualization Of The Primary Language Distribution Within The Svarah Dataset

print("The Distribution Of Primary Languages Within The Svarah Dataset")
print("")

primary_language_distribution = (df['primary_language'].value_counts())
print(f"The Total Number Of Primary Languages : {len(primary_language_distribution)}")
print("")

print("The Ten Most Represented Primary Languages: ")
print("")
display(primary_language_distribution.head(10))
print("")

print("The Coverage Analysis Of Primary Languages: ")

for index, (language, count) in enumerate(primary_language_distribution.head(5).items(), start=1):
    percentage = count / len(df) * 100

    print(
        f"{index}. {language:<15}: "
        f"{count:,} speakers ({percentage:.2f}%)"
    )

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

primary_language_distribution.head(15).plot(
    kind='barh',
    ax=axes[0],
    color='#9b59b6',
    edgecolor='black',
    alpha=0.75
)

axes[0].set_title(
    'The Ten Most Represented Languages',
    fontweight='bold',
    fontsize=12
)

axes[0].set_xlabel('Frequency Count')
axes[0].set_ylabel('Language')
axes[0].grid(axis='x', alpha=0.3)

top_10_languages = primary_language_distribution.head(10)
remaining_languages = primary_language_distribution[10:].sum()

axes[1].pie(
    list(top_10_languages.values) + [remaining_languages],
    labels=list(top_10_languages.index) + ['Others'],
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize': 9}
)
axes[1].pie(
    list(top_10_languages.values) + [remaining_languages],
    labels=list(top_10_languages.index) + ['Others'],
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize': 9}
)

axes[1].set_title(
    'The Distribution Of Primary Languages',
    fontweight='bold',
    fontsize=12
)

plt.tight_layout()
plt.show()
```

```python
# Performing Statistical Analysis And Visualization Of Geographic Distribution Within The Svarah Dataset

print("The Distribution Of Geographic Locations Within The Svarah Dataset")
print("")

state_distribution = (df['native_place_state'].value_counts())
district_distribution = (df['native_place_district'].value_counts())

print(f"The Total Number Of States: {len(state_distribution)}")
print("")
print("The Ten Most Represented States: ")
print("")
display(state_distribution.head(10))
print("")
print(f"The Total Number Of Districts : {len(district_distribution)}")
print("")
print("The Ten Most Represented Districts: ")
print("")
display(district_distribution.head(10))

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

state_distribution.head(12).plot(
    kind='barh',
    ax=axes[0],
    color='#e74c3c',
    edgecolor='black',
    alpha=0.75
)

axes[0].set_title(
    'The Twelve Most Represented States',
    fontweight='bold',
    fontsize=12
)

axes[0].set_xlabel('Frequency Count')
axes[0].set_ylabel('State')
axes[0].grid(axis='x', alpha=0.3)

district_distribution.head(12).plot(
    kind='barh',
    ax=axes[1],
    color='#2ecc71',
    edgecolor='black',
    alpha=0.75
)

axes[1].set_title(
    'The Twelve Most Represented Districts',
    fontweight='bold',
    fontsize=12
)

axes[1].set_xlabel('Frequency Count')
axes[1].set_ylabel('District')
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Performing Statistical Analysis And Visualization Of Educational Qualification Distribution Within The Svarah Dataset

print("The Distribution Of Educational Qualification Within The Svarah Dataset")
print("")

educational_qualification_distribution = (df['highest_qualification'].value_counts())

print("The Levels Of Educational Qualifications: ")
print("")
display(educational_qualification_distribution)

print("")
print("The Distribution Of The Percentages: ")

educational_qualification_percentage_distribution = (educational_qualification_distribution / len(df) * 100).round(2)

for qualification, percentage in (educational_qualification_percentage_distribution.items()):
    print(f"{qualification:<25}: {percentage}%")

plt.figure(figsize=(12, 6))

educational_qualification_distribution.plot(
    kind='barh',
    color='#f39c12',
    edgecolor='black',
    alpha=0.75
)

plt.title(
    'Distribution Of Educational Qualifications',
    fontweight='bold',
    fontsize=12
)

plt.xlabel('Frequency Count')
plt.ylabel('Educational Qualification')
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```
